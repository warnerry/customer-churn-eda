Далее посчитаем метрики - основные показатели по данному датасету

```python
import pandas as pd 
df=pd.read_csv('/Users/denis/Programming/DA/customer_churn_eda/data/processed/churn_clean.csv')
```

Проверим количество пользователей, которые отказались от покупки 

```python
churn_rate=df['churn_flag'].mean()
print(f'Churn rate: {churn_rate:.1%}')
```

```python
contract_metrics = (df.groupby('Contract')
    .agg(customers=('customerID','count'), churn_rate=('churn_flag','mean'), avg_monthly=('MonthlyCharges','mean'))
    .sort_values('churn_rate', ascending=False))
contract_metrics
```

```python
bins=[0,6,12,24,48,100]
labels=['0-6','7-12','13-24','25-48','49+']
df['tenure_group']=pd.cut(df['tenure'],bins=bins,labels=labels,include_lowest=True)
df.groupby('tenure_group')['churn_flag'].agg(['count','mean'])
```

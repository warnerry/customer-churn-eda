```python
import pandas as pd 
df=pd.read_csv('/Users/denis/Programming/DA/customer_churn_eda/data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

Заметим, что колонка TotalCharges содержит в себе числа, но является строкой, что недопустимо

При приведении к числу, появилось 11 NaN значений

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df['TotalCharges'].isna().sum()
```

Удалим соответственно 11 записей, содержащие NaN по этой строчке

```python
df[df['TotalCharges'].isna()]

df = df.dropna(subset=['TotalCharges']).copy()
len(df)
```

Для метрик удобнее иметь бинарный флаг, приведем его с помощью колонки Churn

```python
df['churn_flag']=df['Churn'].map({'Yes':1,'No':0})
df['churn_flag'].to_frame()
```

Найдем долю ушедших клиентов, получим около 27% оттока

```python
df['churn_flag'].mean()
```

Сохраним новый файл

```python
df.to_csv('/Users/denis/Programming/DA/customer_churn_eda/data/processed/churn_clean.csv', index=False)
```
